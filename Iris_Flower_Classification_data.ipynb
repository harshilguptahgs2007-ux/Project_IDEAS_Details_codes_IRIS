```python
import pandas as pd
from sklearn.datasets import load_iris
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df["Species"] = iris.target
limit1 = 7.5
limit2 = 8.0
total_count = 0
setosa_count = 0
versicolor_count = 0
virginica_count = 0
for index, rowData in df.iterrows():
    if rowData["sepal length (cm)"] > limit1 and rowData["sepal length (cm)"] <= limit2:
        total_count += 1
        if rowData["Species"] == 0:
            setosa_count += 1
        elif rowData["Species"] == 1:
            versicolor_count += 1
        else:
            virginica_count += 1

print("Total Count -->", total_count)
print("Setosa Count -->", setosa_count)
print("Versicolor Count -->", versicolor_count)
print("Virginica Count -->", virginica_count)
```

```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df["Species"] = iris.target
for start in np.arange(0, 3, 0.5):
    end = start + 0.5
    total = 0
    setosa = 0
    versicolor = 0
    virginica = 0

    print("Petal Width Range -->", round(start,1), "to", round(end,1))

    for index, rowData in df.iterrows():
        if rowData["petal width (cm)"] > start and rowData["petal width (cm)"] <= end:
            total += 1

            if rowData["Species"] == 0:
                setosa += 1
            elif rowData["Species"] == 1:
                versicolor += 1
            else:
                virginica += 1

    print("Total Count -->", total)
    print("Setosa -->", setosa)
    print("Versicolor -->", versicolor)
    print("Virginica -->", virginica)
    print("------------------------------")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
df = pd.read_csv("Summarised Data.csv")
df.columns = df.columns.str.strip()
df["Sepal Length"] = df["Sepal Length"].astype(str)

# Remove NaN rows if present
df = df[df["Sepal Length"] != "nan"]
x = np.arange(len(df["Sepal Length"]))
width = 0.25
plt.figure()
plt.bar(x - width, df["Setosa(0)"], width, label="Setosa")
plt.bar(x, df["Versicolor(1)"], width, label="Versicolor")
plt.bar(x + width, df["Virginica(2)"], width, label="Virginica")
plt.xlabel("Sepal Length Range")
plt.ylabel("Count")
plt.title("Species Count by Sepal Length Range")
plt.xticks(x, df["Sepal Length"], rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```
